```python
import numpy as np
from scipy.optimize import differential_evolution
import matplotlib.pyplot as plt
```

```python
def ackley(x):
    arg1 = -0.2 * np.sqrt(0.5 * (x[0] ** 2 + x[1] ** 2))
    arg2 = 0.5 * (np.cos(2. * np.pi * x[0]) + np.cos(2. * np.pi * x[1]))
    return -20. * np.exp(arg1) - np.exp(arg2) + 20. + np.e

bounds = [(-10, 10), (-10, 10)]
```

```python
strategies = ['best1bin', 'best1exp', 'rand1exp', 'randtobest1exp', 'currenttobest1exp', 'best2exp', 'rand2exp', 'randtobest1bin', 'currenttobest1bin', 'best2bin', 'rand2bin', 'rand1bin']
res = {}
for st in strategies:
    result = differential_evolution(ackley, bounds, seed=42, strategy = st)
    res[st] = result.nit
```

```python
plt.figure(figsize=(7,5))

for k, v in res.items():
    plt.plot(v, k, 'bo')
    if v == min(res.values()):
        print(f"strategy = {k}")

```

```python
popsizes = np.arange(5, 100, 5).tolist()
res = {}    
for pop in popsizes:
    result = differential_evolution(ackley, bounds, seed=42, strategy = 'best1bin', popsize = pop)
    res[pop] = result.nit

print(res)
```

```python
plt.figure(figsize=(15,5))
plt.ylim([80,100])

for k, v in res.items():
    plt.bar(k, v,width = 3)
    if v == min(res.values()):
        print(f"popsize = {k}")                                                                                                            
```

```python
mutation = np.arange(0, 2, 0.2).tolist()

res = {}    
for mut in mutation:
    result = differential_evolution(ackley, bounds, seed=42, strategy = 'best1bin', popsize = 15, mutation = mut)
    res[mut] = result.nit

plt.figure(figsize=(7,5))

for k, v in res.items():
    plt.plot(k, v, 'bo')
    if v == min(res.values()):
        print(f"mutation = {k}")        
```

```python
recombination = np.arange(0, 1, 0.1).tolist()

res = {}    
for rec in recombination:
    result = differential_evolution(ackley, bounds, seed=42, strategy = 'best1bin', popsize = 15, mutation = 0, recombination = rec)
    res[rec] = result.nit

for k, v in res.items():
    plt.plot(k, v, 'bo')
    if v == min(res.values()):
        print(f"recombination = {k}")

print(res)
```
