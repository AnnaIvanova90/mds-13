```python
import numpy as np
from scipy.stats import norm, expon, t
import matplotlib.pyplot as plt
import math
from scipy import integrate
```

```python
#нормальное распределение
x_axis = np.arange(-10, 10, 0.001)
plt.figure(figsize=(10, 10))
# Среднее значение = 0, стандартное отклонение = 2.
plt.plot(x_axis, norm.pdf(x_axis,0,2))
plt.plot(x_axis, norm.pdf(x_axis,1,2))
plt.plot(x_axis, norm.pdf(x_axis,1,1))
plt.plot(x_axis, norm.pdf(x_axis,3,3))
plt.plot(x_axis, norm.pdf(x_axis,5,4))
plt.show()
```

```python
#экспоненциальное распределение
plt.figure(figsize=(10, 10))
x_axis = np.linspace(expon.ppf(0.01), expon.ppf(0.99), 100)
plt.plot(x_axis, expon.pdf(x_axis, 0,0.5))
plt.plot(x_axis, expon.pdf(x_axis,0,0.33))
plt.plot(x_axis, expon.pdf(x_axis,0,1))
plt.plot(x_axis, expon.pdf(x_axis,1,2))
plt.plot(x_axis, expon.pdf(x_axis,2,5))
```

```python
#распределение Стьюдента
DF = [1, 2, 10, 15, 100]
plt.figure(figsize=(10, 10))
plt.xlim(-5, 5)
for df in DF:
    x = np.linspace(t.ppf(0.01, df), t.ppf(0.99, df), 100)
    plt.plot(x, t.pdf(x, df), alpha=0.6, label='t pdf')
```

епрерывная случайная величина задана плотностью распределения:$$ f(\xi) = C, \xi \in [0;5];f(\xi) = 0, \xi \notin [0;5]$$Найдите C, математическое ожидание $\xi$ а также вероятность попадания в отрезок $[3.5;7]$, т.е. P(3.5 < $\xi$ < 7).

```python
y = lambda x: 1 if x>=0 and x<=5 else 0
 
C = 1/ (integrate.quad(y, 0, 5)[0])
print(f"значение С равно {C}")

y1 = lambda x: 0.2 * x if x >= 0 and x <= 5 else 0

MX = integrate.quad(y1, 0, 5)
print(f"математическое ожидание равно {MX[0]}")


def F(x1):
    if x1 < 0 : return(0)
    elif x1 >= 0 and x1 <= 5: return(0.2 * x1)
    else: return(1)


print(f"вероятность попадания в заданный отрезок {round(F(7) - F(3.5), 2)}")
```

Задание 3
Известно, что светофор горит 54 секунды красным, 4 секунды жёлтым и 28 секунд зелёным. Посчитайте, какое количество информации несёт сообщение о цвете светофора в текущий момент. А что будет в случае 28, 28 и 28 секунд для каждого цвета?

```python
red = 54
yel = 4
grn = 28
p_red = red / (red + yel + grn)
p_yel = yel / (red + yel + grn)
p_grn = grn / (red + yel + grn)
L = - (p_red * math.log2(p_red) + p_yel * math.log2(p_yel) + p_grn * math.log2(p_grn))
print(f"количество информации в сообщение о текущем состоянии свктофора - {round(L,2)} бит")
```
